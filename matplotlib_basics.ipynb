# John D. Hunter – Creator of Matplotlib

- John D. Hunter (1968–2012) was an American neurobiologist and the original creator of Matplotlib.  
- He grew up in Tennessee and graduated from Princeton University in 1990.  
- He earned his Ph.D. in neurobiology from the University of Chicago in 2004.  
- Hunter developed Matplotlib during his postdoctoral research to visualize brain signal (ECoG) data from epilepsy patients.  
- Matplotlib became the most widely used plotting library in Python.  
- It is a core component of the scientific Python ecosystem alongside NumPy, SciPy, and IPython.  
- The library has been used in major scientific achievements, including NASA’s Phoenix Mars mission and the first black hole image.  
- Hunter was also a founding director of the NumFOCUS Foundation.  
- After his passing in 2012, he received the first Distinguished Service Award from the Python Software Foundation.  
- His legacy continues through the annual John Hunter Excellence in Plotting Contest at the SciPy Conference.


# Introduction to Matplotlib

## What is Matplotlib?

Matplotlib is a comprehensive data visualization library in Python used to create static, animated, and interactive visualizations.  
It provides full control over chart elements and allows users to build high-quality plots suitable for analysis, research, reporting, and presentations.

Matplotlib is the foundational visualization library in the Python ecosystem and serves as the base for higher-level libraries such as Seaborn.

---

## Why Use Matplotlib?

Data analysis is not complete without visualization. While numerical summaries provide insight, visual representation reveals patterns, trends, relationships, and anomalies more clearly.

Matplotlib is used because it:

- Provides complete control over every element of a plot  
- Supports a wide variety of chart types  
- Integrates seamlessly with NumPy and Pandas  
- Allows precise customization for professional reporting  
- Is reliable, stable, and widely adopted in scientific computing  

It enables analysts to move from raw numbers to meaningful visual interpretation.

---

## What is Matplotlib Used For?

Matplotlib is widely used in:

- Exploratory Data Analysis (EDA)  
- Statistical visualization  
- Trend analysis and time-series visualization  
- Comparing categorical variables  
- Displaying distributions and relationships  
- Academic research and scientific publications  
- Business dashboards and performance reporting  

It supports essential visualization types such as:

- Line plots  
- Bar charts  
- Histograms  
- Scatter plots  
- Box plots  
- Subplots and multi-figure dashboards  

---

## Summary

Matplotlib transforms numerical data into clear and interpretable visual representations.  
It is a core tool for anyone working in data analysis, data science, scientific research, or machine learning with Python.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create dataset
np.random.seed(42)

data = {
    "Student": range(1, 51),
    "Math": np.random.normal(75, 10, 50),
    "Science": np.random.normal(70, 12, 50),
    "English": np.random.normal(80, 8, 50),
    "Study_Hours": np.random.randint(1, 6, 50),
    "Gender": np.random.choice(["Male", "Female"], 50)
}

df = pd.DataFrame(data)
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
plt.plot(df["Math"])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df["Math"], label="Math Scores")
plt.title("Math Scores Trend")
plt.xlabel("Student Index")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(df["Math"], label="Math", color="blue")
plt.plot(df["Science"], label="Science", color="green")
plt.plot(df["English"], label="English", color="red")

plt.legend()
plt.title("Subject Comparison")
plt.xlabel("Student Index")
plt.ylabel("Score")
plt.grid(True)

plt.show()
```

## Bar Chart

```python
means = df[["Math","Science","English"]].mean()

plt.figure(figsize=(6,4))
plt.bar(means.index, means.values)
plt.title("Average Score per Subject")
plt.ylabel("Average Score")
plt.show()
```

```python
plt.barh(means.index, means.values)
plt.show()
```

```python
gender_means = df.groupby("Gender")[["Math","Science"]].mean()

x = np.arange(len(gender_means.index))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, gender_means["Math"], width, label="Math", color="red")
plt.bar(x + width/2, gender_means["Science"], width, label="Science")

plt.xticks(x, gender_means.index)
plt.legend()
plt.title("Gender Comparison")
plt.show()
```

## Histogram

```python
plt.figure(figsize=(7,5))
plt.hist(df["Math"], bins=10)
plt.title("Distribution of Math Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(5, 4))
plt.hist(df["Math"], bins=10, alpha=0.5, label="Math", color="red")
plt.hist(df["Science"], bins=10, alpha=0.5, label="Science")
plt.legend()
plt.show()
```

```python
from scipy.stats import gaussian_kde

plt.figure(figsize=(6, 4))

data = np.random.normal(75, 10, 50)

# Histogram normalized
plt.hist(data, bins=10, density=True, alpha=0.5)

# KDE line
kde = gaussian_kde(data)
x_vals = np.linspace(min(data), max(data), 200)
plt.plot(x_vals, kde(x_vals), color='black', linewidth=2)

plt.title("Histogram with Density Line")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

data = {
    "Student": range(1, 51),
    "Math": np.random.normal(75, 10, 50),
    "Science": np.random.normal(70, 12, 50),
    "English": np.random.normal(80, 8, 50),
    "Study_Hours": np.random.randint(1, 6, 50),
}

df = pd.DataFrame(data)
df
```

```python
from scipy.stats import gaussian_kde

x = df["Math"]

kde = gaussian_kde(x)
xs = np.linspace(x.min(), x.max(), 200)

plt.figure(figsize=(8,5))
plt.plot(xs, kde(xs))
plt.title("Kernel Density Estimation - Math Scores")
plt.xlabel("Score")
plt.ylabel("Density")
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(6,5))
plt.imshow(corr, cmap="coolwarm", interpolation="none")
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=45)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
x = df["Study_Hours"]
y = df["Math"]

coeffs = np.polyfit(x, y, 2)
poly = np.poly1d(coeffs)

x_line = np.linspace(x.min(), x.max(), 100)

plt.figure(figsize=(8,5))
plt.scatter(x, y)
plt.plot(x_line, poly(x_line), color="red")
plt.title("Polynomial Regression Fit")
plt.xlabel("Study Hours")
plt.ylabel("Math Score")
plt.show()
```

```python
y_pred = poly(x)
residuals = y - y_pred

plt.figure(figsize=(8,5))
plt.scatter(y_pred, residuals)
plt.axhline(0, color="red")
plt.title("Residual Plot")
plt.xlabel("Predicted Score")
plt.ylabel("Residual")
plt.show()
```

```python
sorted_scores = np.sort(df["Math"])
cdf = np.arange(1, len(sorted_scores)+1) / len(sorted_scores)

plt.figure(figsize=(8,5))
plt.plot(sorted_scores, cdf)
plt.title("Cumulative Distribution Function (Math)")
plt.xlabel("Score")
plt.ylabel("CDF")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df["Study_Hours"], df["Math"], df["Science"])

ax.set_xlabel("Study Hours")
ax.set_ylabel("Math")
ax.set_zlabel("Science")
plt.title("3D Scatter Plot")
plt.show()
```

```python
rolling_mean = df["Math"].rolling(window=5).mean()

plt.figure(figsize=(8,5))
plt.plot(df["Math"], alpha=0.4)
plt.plot(rolling_mean, color="red")
plt.title("Moving Average (Window=5)")
plt.show()
```

```python
subjects = ["Math", "Science", "English"]
means = df[subjects].mean()

angles = np.linspace(0, 2*np.pi, len(subjects), endpoint=False)
values = np.concatenate((means.values, [means.values[0]]))
angles = np.concatenate((angles, [angles[0]]))

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, polar=True)

ax.plot(angles, values)
ax.fill(angles, values, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(subjects)

plt.title("Average Performance Radar")
plt.show()
```
