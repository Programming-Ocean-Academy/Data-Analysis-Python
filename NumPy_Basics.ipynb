# Introduction to NumPy

## What is NumPy?
NumPy (Numerical Python) is a core Python library designed for efficient numerical computation.  
It provides a powerful data structure called the ndarray, which allows you to store and manipulate large collections of numbers in a fast and memory-efficient way.

NumPy is the foundation of most data-related libraries in Python and is widely used in data analysis, scientific computing, machine learning, and artificial intelligence.

---

## Why NumPy?
Python lists are flexible, but they become slow and inefficient when working with large numerical datasets.  
NumPy solves this problem by offering:

- Much faster computations than Python lists  
- Efficient memory usage  
- Built-in mathematical and statistical functions  
- Vectorized operations without explicit loops  
- A solid foundation for libraries like Pandas, Scikit-Learn, TensorFlow, and PyTorch  

In data analysis, speed, clarity, and reliability are essential, and NumPy provides all of them.

---

## Where is NumPy Used?
NumPy is used across many real-world data and AI workflows, including:

- Data analysis and data preprocessing  
- Data cleaning and transformation  
- Statistical calculations and aggregations  
- Machine learning pipelines  
- Scientific and engineering computations  
- Artificial intelligence and deep learning foundations  

If you work with numerical data in Python, NumPy is an essential tool.

---

## Key Idea to Remember
NumPy turns numerical data into fast, structured, and analyzable objects, making Python suitable for serious data analysis and AI work.

## Creating NumPy Arrays

```python
data = [10, 20, 30, 40]
arr = np.array(data)

arr
```

```python
type(arr)
```

## Key properties

```python
print(f"{arr.ndim}")     # number of dimensions
print(f"{arr.shape}")    # rows, columns
print(f"{arr.size}")     # total elements
print(f"{arr.dtype}")    # data type
```

## Mini Analytical Lab

```python
scores = np.array([
    [80, 90, 75],
    [85, 88, 92],
    [70, 65, 60],
    [95, 98, 100]
])
```

```python
# Average score per student
scores.mean(axis=1)
```

```python
# Average score per subject
scores.mean(axis=0)
```

```python
# Students with average > 85
avg_scores = scores.mean(axis=1)
```

```python
avg_scores[avg_scores > 85]
```
