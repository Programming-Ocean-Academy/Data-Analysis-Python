# Introduction to Python

Python is a high-level, interpreted programming language known for its **simplicity**, **readability**, and **power**.  
It was designed with a clear philosophy: *code should be easy to read and easy to write*.  
Because of this, Python has become one of the most popular programming languages in the world.

Python supports multiple programming paradigms, including:
- Procedural programming  
- Object-Oriented Programming (OOP)  
- Functional programming  

Its clean syntax makes it an ideal first language for beginners, while its rich ecosystem of libraries makes it a powerful tool for professionals.

---

# Founder of Python

Python was created by, a Dutch software engineer.  
He started developing Python in the late 1980s and released the first version in **1991**.

Guidos main goal was to create a language that:
- Emphasizes code readability  
- Reduces complexity  
- Allows developers to focus on solving problems rather than syntax  

He is often referred to as *Pythons Benevolent Dictator For Life (BDFL)* for his long-term leadership and guidance of the languages design philosophy.

## Dictionaries

```python
student = {
    "name": "Steve",
    "age": 21,
    "grade": "A"
}

print(student["name"])
print(student["age"])
```

```python
student["age"] = 22
student["city"] = "New York"

print(student)
```

## Looping through dictionaries

```python
for key, value in student.items():
    print(key, ":", value)
```

## Mini Project

```python
def check_grade(name, score):
    if score >= 90:
        grade = "A"
    elif score >= 75:
        grade = "B"
    else:
        grade = "C"

    return f"{name}'s grade is {grade}"

print(check_grade("Ahmed", 88))
print(check_grade("Sara", 95))
```
