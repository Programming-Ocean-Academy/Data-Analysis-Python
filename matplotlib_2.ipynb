## Data Prepration and Cleaning

```python
# Ensure numeric columns
num_cols = ["Units_Sold","Unit_Price_USD","Discount_Percent",
            "Revenue_USD","Customer_Rating","Delivery_Days"]
```

```python
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
df["Units_Sold"].median()
```

```python
# ===============================
# 1) Units_Sold  Median
# ===============================
units_median = df["Units_Sold"].median()

df["Units_Sold"] = df["Units_Sold"].fillna(units_median)
```

```python
df["Units_Sold"].isna().sum()
```

```python
int(df["Units_Sold"].isna().sum())
```

```python
# ===============================
# 2) Discount_Percent  0
# ===============================
df["Discount_Percent"] = df["Discount_Percent"].fillna(0)
```

```python
df["Units_Sold"] * df["Unit_Price_USD"] *(1 - df["Discount_Percent"] / 100)
```

```python
# ===============================
# 3) Revenue_USD  Recalculate if missing
# ===============================
df["Revenue_USD"] = df["Revenue_USD"].fillna(
    df["Units_Sold"] *
    df["Unit_Price_USD"] *
    (1 - df["Discount_Percent"] / 100)
)
```

```python
# ===============================
# 4) Customer_Rating  Median
# ===============================
rating_median = df["Customer_Rating"].median()

df["Customer_Rating"] = df["Customer_Rating"].fillna(rating_median)
```

```python
# ===============================
# 5) Delivery_Days  Median
# ===============================
delivery_median = df["Delivery_Days"].median()
df["Delivery_Days"] = df["Delivery_Days"].fillna(delivery_median)

# Final Check
print("Remaining Missing Values:")
print(df.isna().sum())
```

## Stacked Bar: Season  Category

```python
# ==============================
# STACKED BAR
# ==============================

pivot = df.pivot_table(index="Season",
                       columns="Product_Category",
                       values="Revenue_USD",
                       aggfunc="sum",
                       fill_value=0)

bottom = np.zeros(len(pivot))

plt.figure(figsize=(8,6))

for col in pivot.columns:
    plt.bar(pivot.index, pivot[col], bottom=bottom, label=col)
    bottom += pivot[col]

plt.legend()
plt.title("Revenue by Season and Category")
plt.ylabel("Revenue")
plt.show()
```

1 Winter overwhelmingly drives total revenue, and this dominance is mainly fueled by Computer sales, which form the largest portion of the winter bar.

2 In all seasons, the Computer category consistently contributes the highest share, confirming it as the core revenue driver year-round.

3 Audio performs moderately, with noticeable increases in Winter, suggesting seasonal demand effects.

4 Accessories contribute the smallest portion across all seasons, indicating they are supplementary products rather than primary revenue generators.

```python
# =========================================
# LINE CHART: Units Sold Over Time
# =========================================

import matplotlib.pyplot as plt
import pandas as pd

# Ensure Date is datetime
df["Date"] = pd.to_datetime(df["Date"])

# Sort by Date
df_sorted = df.sort_values("Date")

# Optional: Aggregate by Date (if multiple orders per day)
daily_units = df_sorted.groupby("Date")["Units_Sold"].sum()

# Plot
plt.figure(figsize=(10,5))
plt.plot(daily_units.index, daily_units.values, marker='o')

plt.title("Units Sold Over Time")
plt.xlabel("Date")
plt.ylabel("Units Sold")
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```

1 Sales show high volatility throughout the year, with frequent ups and downs rather than a smooth trend.

2 There are several sharp spikes, especially toward the end of the year, including a major peak around ~500 units  likely seasonal or promotional effects.

3 The final months (NovDec period) show significantly higher sales activity, indicating strong year-end demand.

4 Earlier months display mostly moderate-to-low sales levels, suggesting growth momentum builds later in the year.

Overall, the trend suggests strong seasonality with a powerful year-end surge in units sold.
